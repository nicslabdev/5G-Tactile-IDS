```python
import pandas as pd

# Define the path to your CSV file
csv_file_path = "../data/all_data_vicomtech.csv"

# Read the CSV file while skipping the first 4 metadata lines
df = pd.read_csv(
    csv_file_path,
    skiprows=3,  # Skip the first 4 lines of metadata
    parse_dates=['_time']  # Parse the '_time' column as datetime
)

# Print the DataFrame to verify
df.head()
```

```python
# Pivot the DataFrame to have _time as index and _field as columns
pivot_df = df.pivot_table(
    index='_time',
    columns='_field',
    values='_value',
    aggfunc='first'  # Use 'first' to handle duplicates if any
).reset_index()

# Print the DataFrame to verify
pivot_df.to_csv("../data/all_data_vicomtech_sorted.csv")
stats = pivot_df.describe().loc[['min', 'max', 'mean']]

# Print the statistics
print(stats)
```
